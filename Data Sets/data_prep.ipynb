```python
def format_conversation(text):
    human_turns = text.split('Human: ')[1:]
    formatted_conversation = '\n\n'.join(human_turns)

    assistant_turns = formatted_conversation.split('Assistant: ')[1:]
    assistant_responses = [response.split('\n')[0] for response in assistant_turns]

    formatted_assistant_responses = '\n'.join([f'Assistant: {response}\n' for response in assistant_responses])
    return formatted_assistant_responses

# Apply the function to the 'chosen' column
df['chosen_formatted'] = df['chosen'].apply(format_conversation)

# Apply the function to the 'rejected' column
df['rejected_formatted'] = df['rejected'].apply(format_conversation)

# Display the modified DataFrame
print(df[['chosen_formatted', 'rejected_formatted']])
```

```python
# Drop multiple columns
df.drop(columns=['chosen', 'rejected'], inplace=True)
```

```python
import pandas as pd

# Assuming 'df' is your existing DataFrame
data = df['chosen_formatted'].tolist()
types = df['chosen_type'].tolist()

new_df = pd.DataFrame({'data': data, 'type': types})

# Display the new DataFrame
print(new_df)
```

```python
# Assuming 'df' is your existing DataFrame and 'new_df' is the DataFrame created earlier
rejected_data = df['rejected_formatted'].tolist()
rejected_types = df['rejected_type'].tolist()

new_df = new_df.append(pd.DataFrame({'data': rejected_data, 'type': rejected_types}), ignore_index=True)

# Display the updated new DataFrame
print(new_df)
```

```python
# Function to split the 'Data' column into multiple cells and assign the 'type' value
def split_and_assign(row):
    sentences = row['data'].split('Assistant: ')[1:]
    types = [row['type']] * len(sentences)
    return pd.DataFrame({'Assistant': sentences, 'type': types})

# Apply the function to each row and concatenate the results
split_data = pd.concat(df.apply(split_and_assign, axis=1).tolist(), ignore_index=True)

# Display the result
split_data.tail()
```
